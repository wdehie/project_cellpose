import package

```python
import os
import numpy as np
import matplotlib.pyplot as plt

from cellpose import models, io
from cellpose.io import imread
import tifffile as tiff
from cellpose.utils import outlines_list

from IPython.display import display, clear_output
```

```python
import glob
import napari
import time
```

import input directory

```python
input_dir = input("Enter input image path: ")
input_dir = str(input_dir)
print("image path:", input_dir)
```

load image

```python
#loading multiple image
'''patterns = ["*.tif", "*.tiff", "*.png", "*.jpg"]

files = []
for p in patterns:
    files.extend(glob.glob(os.path.join(input_dir, p)))

print(f"Found {len(files)} images")
'''
#still single image
imgs = imread(input_dir)
print("check for image shape:", imgs.shape)
```

load model

```python
start_time = time.time()

model = models.CellposeModel(gpu=False)
masks, flows, styles = model.eval(
    imgs
)

end_time = time.time()
print(f"Segmentation complete: {end_time - start_time:.2f} seconds")
```

view mask

```python
viewer = napari.Viewer()

viewer.add_image(imgs)
viewer.add_labels(masks, opacity=0.4)

napari.run()
```

save mask

```python
mask_root = input("Enter data directory: ")

name = os.path.splitext(os.path.basename(input_dir))[0]

output_dir = os.path.join(mask_root, f"{name}_mask")

os.makedirs(output_dir, exist_ok=True)

print("Mask directory created:", output_dir)

name = os.path.splitext(os.path.basename(input_dir))[0]

mask_path = os.path.join(output_dir, f"mask.tiff")

tiff.imwrite(mask_path, masks)

print("Mask saved to:", mask_path)
```
